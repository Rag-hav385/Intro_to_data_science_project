```python
import os
from tqdm import tqdm
import pandas as pd
```

```python
s = list(set([i for i in os.listdir() if i[-3:] == 'csv']))
s
```

```python
datas = []

for file in tqdm(s):
    data = pd.read_csv(file)
    datas.append(data)
    
pd.concat(datas , axis = 0)
```

```python
import re

from_to_values = []

for file_name in s:
    # Use regular expression to extract 'from' and 'to' values
    match = re.search(r'(\d{4}-\d{2}-\d{2})_[^_]+_[^_]+_(\w+)_(\w+)_', file_name)
    
    if match:
        from_value = match.group(2)
        to_value = match.group(3)
        from_to_values.append((from_value, to_value))

print(from_to_values)
```

```python
unique_values = []

for i in from_to_values:
    unique_values.append(i[0])
    unique_values.append(i[1])
    

set(unique_values)
    
```
